```python
import urllib.request
url = 'ftp://ftp.ncbi.nlm.nih.gov/genomes/archive/old_genbank/Bacteria/Mycoplasma_genitalium_uid97/L43967.fna'
urllib.request.urlretrieve(url, 'L43967.fna')
```

```python
a = open("L43967.fna","r")

while True:
    line = a.readline()
    print(line)
    if not line: 
        break
a.close()
```

```python
a = open("L43967.fna","r")
lines = []
while True:
    line = a.readline()
    line = line.replace("\n", "")
    lines.append(line)
    if not line: 
        break
a.close()
```

```python
dna = ""
for line in lines[1: ]:
    dna = dna + line
    
print(dna)
```

```python
import re
```

```python
matches = re.finditer(r"ATG", dna) 
starts = []
for m in matches:
    starts.append(m.start())

print(len(starts))
```

```python
matches = re.finditer(r"TAA|TAG|TGA", dna) 
stops = []
for m in matches:
    stops.append(m.end())
    
print(len(stops))
```

```python
# remove stops that are before the first start
stops = [stop for stop in stops if stop > starts[0]]

```

```python
# remove starts that are after the last stop
starts = [start for start in starts if start < stops[-1]]

```

```python
fasta_file = open("assignment2.fasta", "a")
i = 1
for start in starts:
    start = starts[0]
    stop = stops[0]
    #print("start at " + str(start) + " stop at " + str(stop))
    #print(">sequence_" + str(i))
    #print(dna[start:stop])
    fasta_file.write(">sequence_" + str(i) + "\n" + dna[start:stop] + "\n")
    # remove starts that are before the last stop
    starts = [start for start in starts if start > stop - 1]
    if(len(starts) == 0):
        break
    else:
        # remove stops that are before the next start (with 6 nucleotide buffer)
        stops = [stop for stop in stops if stop > starts[0] + 5]
        i = i + 1
fasta_file.close()
#for i in list(range(len(starts))):
#    print(">sequence_" + str(i + 1))
#    print(dna[starts[i]:ends[i]])
    
```
